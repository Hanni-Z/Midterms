```python
import cv2
import numpy as np

# 1. Load Image
# (Replace with 'oil_painting_result.jpg' to frame your previous work)
filename = 'famileh.jpg' 
img = cv2.imread(filename)

if img is None:
    print(f"Error: Could not load {filename}")
    exit()

inner_thickness = 5
inner_color = [0, 215, 255] # Gold/Yellow (BGR)

step_1_img = cv2.copyMakeBorder(
    img,
    inner_thickness, inner_thickness, inner_thickness, inner_thickness,
    cv2.BORDER_CONSTANT,
    value=inner_color
)

# Step B: Add a thick outer border around Step A (e.g., White or Off-White)
outer_thickness = 40
outer_color = [92, 133, 165]  

fancy_frame = cv2.copyMakeBorder(
    step_1_img, # Notice we pass the result of Step A here
    outer_thickness, outer_thickness, outer_thickness, outer_thickness,
    cv2.BORDER_CONSTANT,
    value=outer_color
)

h, w = img.shape[:2]
view_h = 400
aspect_ratio = w / h
view_w = int(view_h * aspect_ratio)


total_fancy_thick = inner_thickness + outer_thickness
show_fancy = cv2.resize(fancy_frame, (view_w + total_fancy_thick*2, view_h + total_fancy_thick*2))

cv2.imshow("Fancy Double Frame", show_fancy)

# SAVE the fancy frame result
cv2.imwrite('framed_result.jpg', fancy_frame)
print("Saved 'framed_result.jpg'")

cv2.waitKey(0)
cv2.destroyAllWindows()
```
