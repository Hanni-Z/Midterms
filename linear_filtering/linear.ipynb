```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv.imread("anime1.jpg")

# Check if image loaded successfully
if image is None:
    print("Error: Image not found")
    exit()

# Create the figure with 3 subplots
fig, ax = plt.subplots(1, 3, figsize=(16, 8))
fig.tight_layout()

# --- 1. Original Image ---
# Convert BGR to RGB for matplotlib
ax[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')
ax[0].axis('off')  # Optional: Hide axes for cleaner look

# --- 2. Sharpening Kernel 1 ---
kernel_sharpening = np.array([[-1, -1, -1],
                              [-1,  9, -1],
                              [-1, -1, -1]])

sharpened = cv.filter2D(image, -1, kernel_sharpening)

ax[1].imshow(cv.cvtColor(sharpened, cv.COLOR_BGR2RGB))
ax[1].set_title('Sharpened Kernel Image')
ax[1].axis('off')

# --- 3. Sharpening Kernel 2 ---
kernel_sharpening_2 = np.array([[-1, -1, -1],
                                [-1, 10, -1],
                                [-1, -1, -1]])

sharpened_2 = cv.filter2D(image, -1, kernel_sharpening_2)

ax[2].imshow(cv.cvtColor(sharpened_2, cv.COLOR_BGR2RGB))
ax[2].set_title('Sharpened Kernel Image 2')
ax[2].axis('off')

# --- Save and Show ---
output_filename = 'sharpening_output.png'
plt.savefig(output_filename, bbox_inches='tight') # Save the figure
print(f"Comparison saved to {output_filename}")

plt.show()
```
