```python
import cv2
import numpy as np

img = cv2.imread('anime.jpg')

if img is None:
    print("Error: Image not found.")
    exit()

oil_painting = cv2.xphoto.oilPainting(img, 7, 1)


output_filename = 'oil_painting_result.jpg'
success = cv2.imwrite(output_filename, oil_painting)

if success:
    print(f"Success! Image saved as {output_filename}")
else:
    print("Failed to save the image.")

#cv2.imshow('Oil Painting', oil_painting)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
